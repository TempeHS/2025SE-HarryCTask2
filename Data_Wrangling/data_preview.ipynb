### Graphical Data Representation

- Graphs that provides insights into the data which will inform future decisions surrounding model design and wrangling approaches

```python
# plot a line graph 
data_frame.plot()
```

- As you can see the features are varied in scale at the moment, so they will need to be scaled and there is a large outlier around row 8200

```python
# Plot a histogram of a column
plt.hist(data_frame['km_from_cbd'])
plt.title(f"Histogram of {data_frame['km_from_cbd'].name}")
plt.ylabel('Count')
plt.xlabel(f'{data_frame["km_from_cbd"].name}')
plt.show()
```

This is a scatter plot used to graphically represent the relationship between an input feature and the target

```python
# Scatter plot 2 columns to see the relationship
plt.scatter(data_frame['num_bed'], data_frame['Target'])
plt.title(f"Scatter of {data_frame['num_bed'].name} against {data_frame['Target'].name}")
plt.ylabel(f'{data_frame['Target'].name} Data')
plt.xlabel(f'{data_frame['num_bed'].name} Data')
plt.show()
```

- As you can see there is very little relationship between single features and target data

Below is a scatter plot of 2 features to better visualise the relationship

```python
# Scatter plot multiples columns to see the relationship
x_plot = ['num_bed', 'km_from_cbd']
for col in x_plot:
    plt.scatter(data_frame[col], data_frame['Target'], marker='x')
plt.title(f"Scatter of {*x_plot,} against {data_frame['Target'].name}")
plt.ylabel(f'{data_frame['Target'].name} Data')
plt.xlabel(f'{*x_plot,} Data')
plt.show()
```

```python
# Scatter plot 2 columns in separate charts with a shared y-axis
fig, (ax1, ax2) = plt.subplots(1,2, sharey=True)
plt.suptitle(f"Scatter of {data_frame['num_bed'].name} and {data_frame['km_from_cbd'].name} against {data_frame['Target'].name}")
ax1.set_ylabel(f'{data_frame['Target'].name} Data')

ax1.scatter(data_frame['num_bed'], data_frame['Target'])
ax1.set_xlabel(f'{data_frame['num_bed'].name} Data')

ax2.scatter(data_frame['km_from_cbd'], data_frame['Target'])
ax2.set_xlabel(f'{data_frame['km_from_cbd'].name} Data')

plt.show()
```

- Below is a 3D scatter plot, providing an extra dimension to see if there is a relationship between 2 features and the target

```python
# 3D Scatter plot 3 columns to see the relationship

x_plot = ['num_bed', 'km_from_cbd']

fig = plt.figure()
plt.suptitle(f"3D Scatter of {*x_plot,} against {data_frame['Target'].name}")
ax = fig.add_subplot(111, projection='3d')

ax.scatter(data_frame[x_plot[0]], data_frame[x_plot[1]], data_frame['Target'], color='blue')

x1_range = np.linspace(data_frame[x_plot[0]].min(), data_frame[x_plot[0]].max())
x2_range = np.linspace(data_frame[x_plot[1]].min(), data_frame[x_plot[1]].max())
X1_grid, X2_grid = np.meshgrid(x1_range, x2_range)


ax.set_xlabel('Feature 1')
ax.set_ylabel('Feature 2')
ax.set_zlabel('Target')

plt.show()
```

- From the amount of scatter plots, it is clear that the relationship between individual features and the targe is minimal, and requires multiple features to recognise a relationship
