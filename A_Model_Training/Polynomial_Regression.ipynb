### Model Training - Linear Regression

Linear regression assumes the data has a linear pattern. If your data does not have a linear pattern you can consider a polynomial regression algorithm that provides a multi (poly) coefficient linear relationship between an independent variable and a dependent variable to predict the outcome of price. We will now Train a Polynomial regression algorithm on the features you have engineered to see how closely we can predict the house prices based off the given and created features

- For our context, it is clear more terms with higher coefficients will required, as a simple linear equation cannot potray the relationship between our features and the target.

#### Step 1

Load the required dependencies:

```python
# Import frameworks
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use('../style_Matplotlib_charts.mplstyle')
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
import pickle
```

#### Step 2

Store the data columns for Polynomial Regression

```python
# Create a data frame from the csv file
training_data = pd.read_csv('training_data/A_training_data_new_2.csv')

x_name = ['value_score', 'tot_rooms', 'ds_float', 'suburb_lat', 'suburb_lng']
y_name = 'Target'
x = np.array(training_data[x_name])
y = np.array(training_data[y_name])
```

#### Step 3

Plot the feature `x` and target `y` data on a graph using the column headings as the graph labels.

```python
if(len(x_name) == 1):
    plt.scatter(x, y, marker='x', c='r')
    plt.title("Predicting House Prices")
    plt.ylabel(f'Training {y_name}')
    plt.xlabel(f'Training {x_name[0]}')
else:
    fig,ax=plt.subplots(1,len(x_name),figsize=(12,3))
    for i in range(len(ax)):
        ax[i].scatter(x[:,i],y, label = 'target')
        ax[i].set_xlabel(x_name[i])
    ax[0].set_ylabel("Target"); ax[0].legend();
    fig.suptitle("Predicting House Prices")
plt.show()
```

#### Step 4

Then save an instance of PolynomialFeatures, adjust the 'degree' parameter to suit your model and create the feature using the data in 'x'. (Note a degree of 1 is linear regression, each extra degree adds a coefficient).

Then use the [Scikit-learn](https://scikit-learn.org/stable/) library to instantiate a Machine Learning model class and store it in the `my_model` variable as an object.

```python
poly = PolynomialFeatures(degree=6, include_bias=False)
poly_features = poly.fit_transform(x)

# Create the model
my_model = LinearRegression()
# Fit the model to the data
my_model.fit(poly_features, y)
```

- The higher the degree the longer it takes to train the model. This model of degree 8 can take upwards of a minute

#### Step 5

Plot the features, targets and model (linear regression).

```python
y_pred = my_model.predict(poly_features)
if(len(x_name) == 1):
    plt.scatter(x, y_pred)
    plt.scatter(x, y, marker='x', c='r')
    plt.title("Predicting House Prices")
    plt.ylabel(f'Training {y_name}')
    plt.xlabel(f'Training {x_name[0]}')
else:
    fig,ax=plt.subplots(1,len(x_name),figsize=(12,3),sharey=True)
    for i in range(len(ax)):
        ax[i].scatter(x[:,i],y, label = 'target')
        ax[i].set_xlabel(x_name[i])
        ax[i].scatter(x[:,i],y_pred,color="orange", label = 'predict')
    ax[0].set_ylabel("Target"); ax[0].legend();
    fig.suptitle("Predicting House Prices")
plt.show()
```

#### Step 6

Save the model for testing and validation, since importing the testing and validation for polynomial regression models does not work, testing and validation will be done here

```python
# save the model to disk
filename = 'A_PR_model_v6.sav'
pickle.dump(my_model, open(filename, 'wb'))
```

#### Step 7

Load the polynomial regression model used for testing

```python
filename = 'A_PR_model_v6.sav'
model_C = pickle.load(open(filename, 'rb'))
```

Opening and parsing the test data CSV file, with the specific input features collected

```python
mx_col = ['value_score', 'tot_rooms', 'ds_float', 'suburb_lat', 'suburb_lng']
testing_data = pd.read_csv('A_testing_data_new_2.csv')
mx_test = np.array(testing_data[mx_col])
my_test = np.array(testing_data['Target'])
```

#### Step 8

- Individually plotting test data features, targets and predictions from the model, on a graph for each input feature, with the purpose of visually evaluating the accuracy of the model

```python
# Recreate the PolynomialFeatures instance with the same degree used during training
poly = PolynomialFeatures(degree=6, include_bias=False)

# Transform the test data using PolynomialFeatures
poly_features_test = poly.fit_transform(mx_test)

# Make predictions using the transformed test data
my_pred = model_C.predict(poly_features_test)

# Plot predictions and targets vs original features
fig, ax = plt.subplots(1, len(mx_col), figsize=(12, 3), sharey=True)
for i in range(len(ax)):
    ax[i].scatter(mx_test[:, i], my_test, label='target')
    ax[i].set_xlabel(mx_col[i])
    ax[i].scatter(mx_test[:, i], my_pred, color="orange", label='predict')
    ax[i].set_ylabel("Target")
    ax[i].legend()
fig.suptitle("Polynomial House Price Predictions")
plt.show()
```

```python
test_score = my_model.score(poly_features_test, my_test)
print(f'Training data score: {test_score}')
```

- This score is lower both the initial multivariable score, so it is clear that polynomial regression is not the optimal algorithm, in addition to the increased amount of computational power it requires. No further testing is required, and will not be used on the second (B) model
