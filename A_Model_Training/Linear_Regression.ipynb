### Model Training - Linear Regression

Linear regression is an algorithm that provides a linear relationship between an independent variable and a dependent variable to predict the outcome of future events. We will now Train a Linear regression algorithm on the features you that we're engineered to test how a single input feature can be used to predict the price of a house

#### Step 1

Load the required dependencies:

```python
# Import frameworks
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use('../style_Matplotlib_charts.mplstyle')
from sklearn.linear_model import LinearRegression
import pickle
```

#### Step 2

Store the data columns for Linear Regression.

```python
# Create a data frame from the csv file
training_data = pd.read_csv('training_data/A_training_data_new_1.csv')

x_name = 'prop_cbd'
y_name = 'Target'
x = np.array(training_data[x_name]).reshape(-1, 1)
y = np.array(training_data[y_name])
```

#### Step 3

Plot the feature `x` and target `y` data on a graph using the column headings as the graph labels.

```python
# Plot the data points
plt.scatter(x, y, marker='x', c='r')
# Set the title
plt.title("Predicting House Price")
# Set the y-axis label
plt.ylabel(f'Training {training_data[y_name].name}')
# Set the x-axis label
plt.xlabel(f'Training {training_data[x_name].name}')
plt.show()
```

- As you can see there is a relationship simply by the property size * distance from the CBD, and it resembles that off a negative exponential graph. Thus it is clear that a single variable linear regression will be extremely innacurate when making a prediction.

#### Step 4

Use the [Scikit-learn](https://scikit-learn.org/stable/) library to instantiate a Machine Learning model class and store it in the `my_model` variable as an object.

```python
# Create the model
my_model = LinearRegression()
# Fit the model to the data
my_model.fit(x, y)
```

#### Step 5

Plot the features, targets and model (linear regression).

```python
y_pred = my_model.predict(x)
plt.scatter(x, y, marker='x', c='r')
plt.scatter(x, y_pred)
plt.title("Predicting House Price")
plt.xlabel(f'Training {training_data[x_name].name}')
plt.ylabel(f'Training {training_data[y_name].name}')
plt.show()
```

- As you can see the linear regression line does show a small relationship however the cost of the model (distance of line from data points) is very large. Indicating linear regression is not an appropriate Machine Learning Model and More Features will need to be provided
